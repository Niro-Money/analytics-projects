```python
import pandas as pd
```

```python
def score_bkt(row, var):
    if row[var] <= 600:
        return '<=600'
    elif row[var] <= 620:
        return '601-620'
    elif row[var] <= 640:
        return '621-640'
    elif row[var] <= 660:
        return '641-660'
    elif row[var] <= 680:
        return '661-680'
    elif row[var] <= 700:
        return '681-700'
    elif row[var] <= 720:
        return '701-720'
    elif row[var] <= 740:
        return '721-740'
    elif row[var] <= 760:
        return '741-760'
    elif row[var] <= 780:
        return '761-780'
    elif row[var] <= 800:
        return '781-800'
    else:
        return '800+'
```

```python
data = pd.read_csv(r"C:\Users\T14s\Desktop\offers\405. Finbud DS1(Oct2023)\Output\405. Finbud DS1 Oct2023 300923.csv",low_memory=False)
data = data[data['dec_reason']=='NOT DECLINED']
```

```python
data['naps_tu_thirtyplus']
```

```python
data['napsthirtyplus_bkt']= data.apply(score_bkt,var='naps_tu_thirtyplus',axis=1)
```

```python
data['napsthirtyplus_bkt'].value_counts()
```

```python
data['max_tenor'].value_counts()
```

```python
df_ann = pd.pivot_table(
        data,
        values="Member Reference",
        index=["napsthirtyplus_bkt"],
        columns=["max_tenor"],
        aggfunc=["count"],
        fill_value=0,
        margins=True,
        margins_name="Total",
    ).rename(columns={'count':'phone_numbers(count)'})
df_ann.head(10)
```

```python
data['SYSTEM_RATEOFINT']=data['SYSTEM_RATEOFINT'].round(2)
```

```python
df_ann1 = pd.pivot_table(
        data,
        values="SYSTEM_LOAN_OFFER",
        index=["napsthirtyplus_bkt"],
        columns=["max_tenor"],
        aggfunc=["mean"],
        fill_value=0,
        # margins=True,
        # margins_name="Total",
    ).rename(columns={'mean':'system_loan_offer(mean)'})
df_ann1.head(10)
```

```python
df_ann2 = pd.pivot_table(
        data,
        values="SYSTEM_RATEOFINT",
        index=["napsthirtyplus_bkt"],
        columns=["max_tenor"],
        aggfunc=["mean"],
        fill_value=0,
    
        # margins=True,
        # margins_name="Total",
    ).rename(columns={'mean':'rate_of_int(mean)'})
df_ann2.head(10)
```

```python
df = pd.concat([df_ann,df_ann1,df_ann2],axis=1,ignore_index=False)
```
