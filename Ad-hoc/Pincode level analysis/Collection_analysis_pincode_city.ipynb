```python
df_oct = pd.read_excel("Oct - State & City.xlsx", sheet_name = 'Sheet1')
df_nov = pd.read_excel("Nov - State & City.xlsx", sheet_name = 'Sheet1')
df_dec = pd.read_excel("Dec - State & City.xlsx", sheet_name = 'Sheet1')
df_jan = pd.read_excel("Jan - State & City.xlsx", sheet_name = 'Sheet1')

master_coll_feb = pd.read_excel("PL Loan Book Feb'24.xlsx", header=2).rename(columns = {'Niro Opportunity ID':'opportunity_id', 'Pin Code':'pin_code', "City":"city", "Oct'23 - POS":"oct_pos", "Nov'23 - POS":"nov_pos", "Dec'23 - POS":"dec_pos", "Jan'24 - POS":"jan_pos"})
```

```python
disbursed_cases = """
    select * from (select 
    la.id as opportunity_id,
    ou.category,
    la.user_id,
    la.principal_amount,
    o.phone_number,
    ba.bank_name,
    CASE
        when la.loan_disbursed_date_by_ops is null then date(la.disbursement_date)
        else date(la.loan_disbursed_date_by_ops)
    end as disbursement_date
    from niro_rds.niro_data.core__public__loan_application la
    left join NIRO_RDS.NIRO_DATA.CORE__PUBLIC__OFFERS o on o.id = la.offer_id
    left join NIRO_RDS.NIRO_DATA.core__public__offer_user_details ou on ou.offer_id = o.id
    left join NIRO_RDS.NIRO_DATA.CORE__PUBLIC__BANK_ACCOUNTS b on b.id = la.bank_account_id
    left join NIRO_RDS.NIRO_DATA.CORE__PUBLIC__BANK_BRANCHES ba on b.ifsc_code = ba.ifsc_code    
    where la.loan_application_status in ('CLOSED','LOAN_EXECUTED') and la.is_deleted = False and o.is_deleted = False
    )
    where disbursement_date >= '2023-01-01';
"""
total_disbursed_cases = extractorFromQuery().queryExecutor(disbursed_cases)
total_disbursed_cases.shape
```

```python
df_jan['Status'].value_counts()
```

```python
df_oct['oct_bounce'] = 1
df_oct['oct_resolution'] = np.where(df_oct['Status'] == 'PAID', 1,
                            np.where(df_oct['Status'] == 'UNPAID', 0, np.nan))

df_nov['nov_bounce'] = 1
df_nov['nov_resolution'] = np.where(df_nov['Status'] == 'Paid', 1,
                            np.where(df_nov['Status'] == 'Unpaid', 0, np.nan))

df_dec['dec_bounce'] = 1
df_dec['dec_resolution'] = np.where(df_dec['Status'] == 'Paid', 1,
                            np.where(df_dec['Status'] == 'Unpaid', 0, np.nan))

df_jan['jan_bounce'] = 1
df_jan['jan_resolution'] = np.where(df_jan['Status'] == 'Paid', 1,
                            np.where(df_jan['Status'] == 'Unpaid', 0, np.nan))
```

```python
master_data = total_disbursed_cases.merge(master_coll_feb[["opportunity_id", "pin_code", "city", "oct_pos", "nov_pos", "dec_pos", "jan_pos"]], on = 'opportunity_id', how = 'inner')\
                .merge(df_oct[['opportunity_id', 'oct_bounce', 'oct_resolution']], on='opportunity_id', how='left')\
                .merge(df_nov[['opportunity_id', 'nov_bounce', 'nov_resolution']], on='opportunity_id', how='left')\
                .merge(df_dec[['opportunity_id', 'dec_bounce', 'dec_resolution']], on='opportunity_id', how='left')\
                .merge(df_jan[['opportunity_id', 'jan_bounce', 'jan_resolution']], on='opportunity_id', how='left')
```

#### Create aggregate variables

##### 1. Pincode level

```python
df_pin = pd.merge(master_data.groupby(['pin_code'])['opportunity_id'].nunique().rename('no_of_customers'),\
master_data.groupby(['pin_code'])['principal_amount'].max().rename('max_principal_amount'), on='pin_code', how='left')\
.merge(master_data.groupby(['pin_code'])['principal_amount'].min().rename('min_principal_amount'), on='pin_code', how='left')\
.merge(master_data.groupby(['pin_code'])['principal_amount'].mean().rename('mean_principal_amount'), on='pin_code', how='left')\
.merge(master_data.groupby(['pin_code'])['principal_amount'].sum().rename('total_principal_amount'), on='pin_code', how='left')\
.merge(master_data.groupby(['pin_code'])['jan_pos'].sum().rename('total_jan_pos'), on='pin_code', how='left')\
.merge(master_data[master_data.oct_pos.notna()].fillna(0).groupby('pin_code')['oct_bounce'].mean().rename('oct_bounce_per'), on='pin_code', how='left')\
.merge(master_data[master_data.nov_pos.notna()].fillna(0).groupby('pin_code')['nov_bounce'].mean().rename('nov_bounce_per'), on='pin_code', how='left')\
.merge(master_data[master_data.dec_pos.notna()].fillna(0).groupby('pin_code')['dec_bounce'].mean().rename('dec_bounce_per'), on='pin_code', how='left')\
.merge(master_data[master_data.jan_pos.notna()].fillna(0).groupby('pin_code')['jan_bounce'].mean().rename('jan_bounce_per'), on='pin_code', how='left')\
.merge((master_data[master_data['oct_bounce'] == 1].groupby('pin_code')['oct_pos'].sum() / master_data[master_data.oct_pos.notna()].groupby('pin_code')['oct_pos'].sum()).rename('oct_bounce_pos_per'), on='pin_code', how='left')\
.merge((master_data[master_data['nov_bounce'] == 1].groupby('pin_code')['nov_pos'].sum() / master_data[master_data.nov_pos.notna()].groupby('pin_code')['nov_pos'].sum()).rename('nov_bounce_pos_per'), on='pin_code', how='left')\
.merge((master_data[master_data['dec_bounce'] == 1].groupby('pin_code')['dec_pos'].sum() / master_data[master_data.dec_pos.notna()].groupby('pin_code')['dec_pos'].sum()).rename('dec_bounce_pos_per'), on='pin_code', how='left')\
.merge((master_data[master_data['jan_bounce'] == 1].groupby('pin_code')['jan_pos'].sum() / master_data[master_data.jan_pos.notna()].groupby('pin_code')['jan_pos'].sum()).rename('jan_bounce_pos_per'), on='pin_code', how='left')\
.merge(master_data[master_data['oct_bounce'] == 1].groupby('pin_code')['oct_resolution'].mean().rename('oct_resolution_per'), on='pin_code', how='left')\
.merge(master_data[master_data['nov_bounce'] == 1].groupby('pin_code')['nov_resolution'].mean().rename('nov_resolution_per'), on='pin_code', how='left')\
.merge(master_data[master_data['dec_bounce'] == 1].groupby('pin_code')['dec_resolution'].mean().rename('dec_resolution_per'), on='pin_code', how='left')\
.merge(master_data[master_data['jan_bounce'] == 1].groupby('pin_code')['jan_resolution'].mean().rename('jan_resolution_per'), on='pin_code', how='left')\
.merge(master_coll_feb[["pin_code", "city"]].drop_duplicates(), on='pin_code', how='left')
```

```python
df_pin.to_excel('pincode_data.xlsx', sheet_name='pincode_data', index=False)
```
